```python
import cv2
import numpy as np
import Utils

def show(img,name="uname"):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


img = cv2.imread("template.png")#read th template
show(img)

ref = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)# translate to gray
show(ref,"gray")

ref = cv2.threshold(ref,10,255,cv2.THRESH_BINARY_INV)[1]#binary
show(ref,"binary")

kernel = np.ones((3,3),np.uint8)
ref = cv2.morphologyEx(ref,cv2.MORPH_CLOSE,kernel)#close to make better
show(ref,"close")


ref_,ref_contours,hierarchy =cv2.findContours(ref.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)#find out the contours

cv2.drawContours(img,ref_contours,-1,(0,0,255),2)
show(img,"img_with_sontours")
ref_contours = Utils.sort_contours(ref_contours,method = "left-to-right")[0]#sort the contours from right to left to make sure that they are corresponded to index 0 to 9

digitals ={}
for (i,c) in enumerate(ref_contours):
    (x,y,w,h)=cv2.boundingRect(c)
    roi = ref[y:y+h,x:x+w]
    roi = cv2.resize(roi,(57,88))
    
    digitals[i] = roi
    show(digitals[i])

#initialize the kernel
rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT,(9,3))
sqKernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
```

```python

img_card = cv2.imread("card1.png")
#轉為灰度圖,改變大小

high = int((img_card.shape[0])*(300/img_card.shape[1]))
img_card = cv2.resize(img_card,(300,high))
img = img_card
show(img_card)

img_card_gray = cv2.cvtColor(img_card,cv2.COLOR_BGR2GRAY)
show(img_card_gray)
img_card = img_card_gray
#禮帽操作,突出明亮區域
img_card = cv2.morphologyEx(img_card,cv2.MORPH_TOPHAT,rectKernel)
show(img_card)

img_card = cv2.Sobel(img_card,ddepth = cv2.CV_32F,dx= 1,dy= 0,ksize = -1) #ksize = -1 相當於ksize = 3

img_card = cv2.convertScaleAbs(img_card)
show(img_card)

img_card = cv2.morphologyEx(img_card,cv2.MORPH_CLOSE,rectKernel)
show(img_card)

ret,img_card = cv2.threshold(img_card,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)#適合雙峰二分,自動判斷閾值
show(img_card)

img_card = cv2.erode(img_card,rectKernel,iterations=1)
show(img_card)
img_card = cv2.dilate(img_card,rectKernel,iterations=2)
show(img_card)

thresh,cnts,hierarchy = cv2.findContours(img_card.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
# img_ = img
# cv2.drawContours(img_,cnts,-1,(0,0,255),2)
show(img)

loc = []
for i,c in enumerate(cnts):
    x,y,w,h = cv2.boundingRect(c)
    ar = w /float(h)
    if(ar>3 and ar <4):
        if(w>55 and w<65) and ( h>14 and h<22):
            loc.append((x,y,w,h))
    cv2.drawContours(img,cnts,i,(0,0,255),2)
    show(img)
    print(ar,w,h)

loc = sorted(loc,key = lambda x:x[0])

nums_img = []
nums=[]
for i,(gx,gy,gw,gh) in enumerate(loc):
    pic = img_card_gray[gy-5:gy+5+gh,gx-5:gx+gw+5]
    pic = cv2.threshold(pic,100,255,cv2.THRESH_BINARY)[1]
    
    ref_,ref_contours,hierarchy =cv2.findContours(pic.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)#find out the contours
    ref_contours = Utils.sort_contours(ref_contours,method = "left-to-right")[0]
    
    for (i,c) in enumerate(ref_contours):
        (x,y,w,h)=cv2.boundingRect(c)
        roi = pic[y:y+h,x:x+w]
        roi = cv2.resize(roi,(57,88))
        nums_img.append(roi)
        show(roi)
        
        similarity = 0
        index =0
        for i,template in digitals.items():
            res = cv2.matchTemplate(roi,template,cv2.TM_CCOEFF)
            minVal= cv2.minMaxLoc(res)[1]
            if similarity < minVal:
                similarity = minVal
                index = i
        nums.append(index)
print(nums)

        
            
            



```
